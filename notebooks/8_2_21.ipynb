```python
import pandas as pd
import plotly.graph_objs as go
```

```python
arabica_dat = pd.read_csv('../data/arabica_data_cleaned.csv')
arabica_df = pd.DataFrame(arabica_dat)
```

```python
arabica_df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
arabica_df = arabica_df[["Country.of.Origin", "Producer", "Processing.Method"]]
```

```python
arabica_df.columns = ["Country of Origin", "Producer", "Processing Method"]
```

```python
properties = ('Country of Origin', 'Producers', 'Processing Method')
def plot_data(data, properties):
    """

    Args:
        data: the dataframe containing all data
        properties: properties/feature in the data which should be shown in the plots

    Returns:
        barplots showing how many times a feature was represented in the data
    """
    for property in properties:
        feature = [
            go.Bar(
                x= data[property].unique(),
                y= data[property].value_counts()
            )
        ]
        fig_countries = go.Figure(data=feature)
        fig_countries.update_layout(xaxis = dict(title = property), yaxis = dict(title = 'Counts'), title="Arabica Coffee-"+ property, width=800, height=400)
        fig_countries.show()



#Producer = [
#    go.Bar(
#        x= arabica_df["Producer"].unique(),
#        y= arabica_df["Producer"].value_counts()
#    )
#]
#fig_producer = go.Figure(data=Producer)
#fig_producer.update_layout(xaxis = dict(title = 'Producers'), yaxis = dict(title = 'Counts'), title="Producers of Arabica Coffee", width=1000, height=400)
#fig_producer.show()

#Processing = [
#    go.Bar(
#        x= arabica_df["Processing Method"].unique(),
#        y= arabica_df["Processing Method"].value_counts()
#    )
#]
#fig_processing = go.Figure(data=Processing)
#fig_processing.update_layout(xaxis = dict(title = 'Processing methods'), yaxis = dict(title = 'Counts'), title="Processing methods of Arabica Coffee", width=800, height=400)
#fig_processing.show()

#for category in ["Country of Origin", "Producer", "Processing Method"]:
#    graph = [
#        go.Bar(
#            x=arabica_df[category].unique(),
#            y=arabica_df[category].value_counts()
#        )
#    ]
#    fig = go.Figure(data=graph)
#    fig.update_layout(title=('Arabica-'+ category), yaxis=dict(title='Counts'), xaxis=dict(title = category))
#    fig.show()
#
#
```
